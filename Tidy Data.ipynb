```python
import pandas as pd
```

```python
#loading in the csv
df = pd.read_csv('tidy_data_example.csv')
```

```python
#changing symbols that mean nothing to NA
df = df.replace('?', 'NA')
```

```python
df = df.replace('-', 'NA')
df
```

## 5 tables to make: 
    1. person - person.csv
    2. book - book.csv
    3. book-genre - genre.csv
    4. book-author

### 1. Person

```python
person_df = df[['First Name', 'Last Name', 'Date of Birth', 'Library Card Expired?']].copy()
```

```python
person_df = person_df.drop_duplicates()
```

```python
person_df.to_csv('person.csv')
```
